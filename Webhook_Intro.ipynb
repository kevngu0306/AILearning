```python
# Week 2 - Day 5: Webhook Fundamentals

import requests

payload = {"name": "Kevin", "status": "Rich"}
response = requests.post("http://localhost:5000/webhook", json=payload)
print(response.status_code)

userInfo = {"name": "Shine", "status": "Mega Rich"}
response = requests.post("http://localhost:5000/webhook", json=userInfo)
print(response.status_code)
```
